```python
import pandas as pd
```

```python
data = df = pd.read_csv('StudentsPerformance.csv')
```

```python
def separate_math_scores(df):

    # Create two lists
    Scores_above_70 = []
    Scores_seventy_and_below = []

    # Loop through math scores
    for score in df['math score']:

        # Append scores to appropriate list
        if score > 70:
            Scores_above_70.append(score)
        else:
            Scores_seventy_and_below.append(score)

    return Scores_above_70, Scores_seventy_and_below


high_scores, low_scores = separate_math_scores(df)

print("Scores above 70:", high_scores)
print("Scores 70 and below:", low_scores)

print("Number of scores above 70:", len(high_scores))
print("Number of scores 70 and below:", len(low_scores))
```

```python
def compare_math_scores_by_gender(data):

    # Filter rows by gender
    male_students = data[data['gender'] == 'male']
    female_students = data[data['gender'] == 'female']

    # Create lists of math scores
    male_scores = male_students['math score'].tolist()
    female_scores = female_students['math score'].tolist()

    # Compute averages
    male_avg = sum(male_scores) / len(male_scores)
    female_avg = sum(female_scores) / len(female_scores)

    # Print comparison
    print("Average Male Math Score:", round(male_avg, 2))
    print("Average Female Math Score:", round(female_avg, 2))


    if male_avg > female_avg:
        print("Pattern Detected: Male students perform better in math on average.")
    elif female_avg > male_avg:
        print("Pattern Detected: Female students perform better in math on average.")
    else:
        print("Pattern Detected: Both genders perform equally in math.")


compare_math_scores_by_gender(df)
```
