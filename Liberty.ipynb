```python
df['TRANSPORT COST (where customer paid)'] = df['TRANSPORT COST (where customer paid)'].fillna(0)
df['TRANSPORT & OTHER UTILITIES (covered by the enterprise)'] = df['TRANSPORT & OTHER UTILITIES (covered by the enterprise)'].fillna(0)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
def clean_currency(col):
    return col.astype(str).str.replace('NGN ', '', regex=False).str.replace(',', '', regex=False)

df["Book Title(s)"] = df["Book Title(s)"].astype(str)
df["Customer's Name"] = df["Customer's Name"].astype(str)
df["ORDER TYPE"] = df["ORDER TYPE"].astype(str)
df["Gender"] = df["Gender"].astype(str)
df["Destination"] = df["Destination"].astype(str)
df["Month"] = df["Month"].astype(str)

df["Cost Price (in Naira)"] = clean_currency(df["Cost Price (in Naira)"]).astype(float).astype(int)
df["Selling Price (in Naira)"] = clean_currency(df["Selling Price (in Naira)"]).astype(float).astype(int)
df["Profit"] = clean_currency(df["Profit"]).astype(float).astype(int)
df["% Profit"] = df["% Profit"].astype(int)
df["TRANSPORT COST (where customer paid)"] = df["TRANSPORT COST (where customer paid)"].astype(int)
df["TRANSPORT & OTHER UTILITIES (covered by the enterprise)"] = df["TRANSPORT & OTHER UTILITIES (covered by the enterprise)"].astype(int)
```

```python
df
```

```python
books_per_month = df.groupby('Month')['Book Title(s)'].count()
print(books_per_month)

# Optional: plot the results
month_order = ['JULY', 'AUGUST', 'SEPTEMBER']
books_per_month_ordered = books_per_month.reindex(month_order)
books_per_month_ordered.plot(kind='bar', figsize=(8,5))
plt.title('Number of Books Sold Per Month')
plt.ylabel('Number of Books')
plt.xlabel('Month')
plt.xticks(rotation=0)
plt.show()
```

```python
total_profit = df["Profit"].sum()
print(total_profit)
```

```python
month_order = ['JULY', 'AUGUST', 'SEPTEMBER']
monthly_profit = df.groupby('Month')['Profit'].sum().reindex(month_order)
print(monthly_profit)
plt.figure(figsize=(8,5))
monthly_profit.plot(kind='bar')
plt.title("Profit by Month (July - September)")
plt.ylabel("Amount (₦)")
plt.xlabel("Month")
plt.xticks(rotation=0)
plt.show()
```

```python
most_sold_books = df['Book Title(s)'].value_counts().head(5)
print(most_sold_books)

plt.figure(figsize=(20,10))
most_sold_books.plot(kind='bar')
plt.title('Top 5 Most Sold Books')
plt.xlabel('Book Title')
plt.ylabel('Number of Copies Sold')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
books_per_month = df.groupby('Month')['Book Title(s)'].count()
print(books_per_month)
books_per_month.plot(kind = 'pie', figsize=(8,5), autopct='%1.1f%%')
plt.title('Monthly Book Distribution')
plt.ylabel('')
plt.xticks(rotation=0)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
print(gender_counts)
gender_counts.plot(kind = 'pie', figsize=(8,5), autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.xticks(rotation=0)
plt.show()
```

```python
destination_disparity = df['Destination'].value_counts()
print(destination_disparity)
destination_disparity.plot(kind = 'bar', figsize=(8,5))
plt.title('Delivery Distribution')
plt.ylabel('')
plt.xticks(rotation=0)
plt.show()
```

```python
type_dist = df['ORDER TYPE'].value_counts()
print(type_dist)
type_dist.plot(kind = 'pie', figsize=(8,5), autopct='%1.1f%%')
plt.title('Order Type')
plt.ylabel('')
plt.xticks(rotation=0)
plt.show()
```

```python
df['Profit_Made'] = df['Profit'] > 0
profit_status_per_month = df.groupby('Month')['Profit_Made'].value_counts().unstack(fill_value=0)

# Reindex to ensure months are in chronological order
month_order = ['JULY', 'AUGUST', 'SEPTEMBER']
profit_status_per_month = profit_status_per_month.reindex(month_order)

print(profit_status_per_month)

profit_status_per_month.plot(kind='bar', stacked=True, figsize=(10, 6), color=['salmon', 'lightgreen'])
plt.title('Profit Status Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=0)
plt.legend(title='Profit Made', labels=['No Profit', 'Profit'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
book_counts = df['Book Title(s)'].value_counts()
single_purchase_books = book_counts[book_counts == 1]
book_counts_filtered = book_counts.drop(single_purchase_books.index)
other_books_count = 1
book_counts_filtered['Others'] = other_books_count
print(book_counts_filtered)
book_counts_filtered.plot(kind = 'bar', figsize=(20,10))
plt.title('Quantity puchased per Book')
plt.ylabel('Number of Books Purchased')
plt.xlabel('Book Title')
plt.show()
```
