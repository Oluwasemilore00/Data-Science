```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("Data1.csv")
```

```python
print(df)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df[['Item', 'Category']] = df[['Item', 'Category']].astype(str)
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
total_spent = df['Debit'].sum()
print(f"Total Amount Spent: ₦{total_spent:.2f}")
total_received = df['Credit'].sum()
print(f"Total Amount Received: ₦{total_received:.2f}")
```

```python
category_spending = df.groupby('Category')['Debit'].sum().sort_values(ascending=False)
print(category_spending)
```

```python
category_spending.plot(kind='bar', figsize=(8,5))
plt.title("Spending by Category")
plt.ylabel("Total Amount (₦)")
plt.xlabel("Category")
plt.show()
```

```python
daily_spending = df.groupby('Date')['Debit'].sum()
daily_spending.plot(kind='line', figsize=(8,5), marker='o')
plt.title("Daily Spending Trend")
plt.ylabel("Amount (₦)")
plt.xlabel("Date")
plt.show()
```
